# Engineer features

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
X = pd.read_csv('X-train.csv', index_col='PassengerId')
y = pd.read_csv('y-train.csv', index_col='PassengerId')
```

```python
X['FamilySize'] = X['SibSp'] + X['Parch'] + 1
X['Alone'] = (X['FamilySize'] == 1).astype(int)
X['Child'] = (X['Age'] < 16).astype(int)
X['Parent'] = ((X['Age'] >= 16) & (X['Parch'] > 0)).astype(int)
X['Mother'] = ((X['Female'] == 1) & (X['Parent'] == 1)).astype(int)
X['Father'] = ((X['Female'] == 0) & (X['Parent'] == 1)).astype(int)
X['OneParent'] = ((X['Child'] == 1) & (X['Parch'] == 1)).astype(int)
X['TwoParent'] = ((X['Child'] == 1) & (X['Parch'] == 2)).astype(int)
X['SmallFamily'] = (X['SibSp'] == 1).astype(int)
X['MediumFamily'] = ((X['SibSp'] > 1) & (X['SibSp'] <= 2)).astype(int)
X['LargeFamily'] = (X['SibSp'] > 2).astype(int)
X['Baby'] = (X['Age'] <= 1).astype(int)
X['Toddler'] = ((X['Age'] > 1) & (X['Age'] <= 3)).astype(int)
X['YoungChild'] = ((X['Age'] > 3) & (X['Age'] < 13)).astype(int)
X['Teenager'] = ((X['Age'] > 13) & (X['Age'] < 18)).astype(int)

X['BabyOneParent'] = ((X['Age'] <= 1) & (X['Parch'] == 1)).astype(int)
X['ToddlerOneParent'] = ((X['Age'] > 1) & (X['Age'] <= 3) & (X['Parch'] == 1)).astype(int)
X['YoungChildOneParent'] = ((X['Age'] > 3) & (X['Age'] < 13) & (X['Parch'] == 1)).astype(int)
X['TeenagerOneParent'] = ((X['Age'] > 13) & (X['Age'] < 18) & (X['Parch'] == 1)).astype(int)

X['BabyTwoParents'] = ((X['Age'] <= 1) & (X['Parch'] == 2)).astype(int)
X['ToddlerTwoParents'] = ((X['Age'] > 1) & (X['Age'] <= 3) & (X['Parch'] == 2)).astype(int)
X['YoungChildTwoParents'] = ((X['Age'] > 3) & (X['Age'] < 13) & (X['Parch'] == 2)).astype(int)
X['TeenagerTwoParents'] = ((X['Age'] > 13) & (X['Age'] < 18) & (X['Parch'] == 2)).astype(int)

X['YoungAdult'] = ((X['Age'] >= 18) & (X['Age'] < 25)).astype(int)
X['MiddleAged'] = ((X['Age'] >= 25) & (X['Age'] < 40)).astype(int)
X['OldAged'] = ((X['Age'] >= 40) & (X['Age'] < 60)).astype(int)
X['Elderly'] = (X['Age'] >= 60).astype(int)
X['CheapFare'] = (X['Fare'] < 40).astype(int)
```
